WR = [(v/v+m)*R] + [(m/v+m)*C]

Weighted Rating
R = rating for movie

v = no.of votes for movie

m = min votes 

C = mean vote across whole report

```python
C = df['rating'].mean()
C
```

```python
m = int(df['votes'].quantile(0.1))
m
```

```python
def weighted_rating(show, m, C):
    R = show['rating']
    v = show['votes']
    return (v/(v+m) * R) + (m/ (m+v) * C)

df.loc[df.votes>m, 'weighted_rating'] = df.loc[df.votes>m].apply(lambda x: weighted_rating(x,m,C), axis=1)

df['weighted_rating'].fillna(0, inplace=True)
```

```python
def make_clickable_both(urldata): 
    names, urls = urldata.official_sites, urldata.links
    if names=='NA' or urls=='NA':
        return 'NA'
    curr_record_html = ''
    for each_name, each_url in zip(names.split(','), urls.split(',')):
        curr_record_html += f', <a href="{each_url}">{each_name}</a>'
    return curr_record_html[2:]

df['nameurls'] = df[['official_sites', 'links']].apply(make_clickable_both, axis=1) 
```

```python
import string
from rake_nltk import Rake
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import numpy as np
```

```python
# Takes a string and returns an array of its processed words.
def clean_string(s):
    # Remove stopwords and punctuation.
    stop = stopwords.words('english') + list(string.punctuation)
    return [n for n in wordpunct_tokenize(s.lower()) if n not in stop]

def create_soup(x):
    title_importance = 1
    language_importance = 1
    stars_importance = 1
    creators_importance = 1
    genre_importance = 1

    soup = ''
    
    # Keywords from description.
    desc = x['description']
    if desc is not np.nan:
        rake = Rake()
        rake.extract_keywords_from_text(desc)
        desc_soup = ' '.join(list(rake.get_word_degrees().keys()))
        soup = ' '.join(filter(None, [soup, desc_soup]))
    
    # Title.
    title_soup = ' '.join(clean_string(x['title']) * title_importance)
    soup = ' '.join(filter(None, [soup, title_soup]))
    
    # Authors.
    language = x['language']
    if language is not np.nan:
        # I'm trying to not remove punctuation here but to just set all as spaces. I want to retain (Role).
        # Providing it's consistent across entries, this should work.
        language_soup = ' '.join([l.lower().replace(' ', '') for l in language.split(',')] * language_importance)
        soup = ' '.join(filter(None, [soup, language_soup]))
    
    # Authors.
    creators = x['creators']
    if creators is not np.nan:
        # I'm trying to not remove punctuation here but to just set all as spaces. I want to retain (Role).
        # Providing it's consistent across entries, this should work.
        creator_soup = ' '.join([c.lower().replace(' ', '') for c in creators.split(',')] * creators_importance)
        soup = ' '.join(filter(None, [soup, creator_soup]))
        
    # Authors.
    stars = x['stars']
    if stars is not np.nan:
        # I'm trying to not remove punctuation here but to just set all as spaces. I want to retain (Role).
        # Providing it's consistent across entries, this should work.
        stars_soup = ' '.join([s.lower().replace(' ', '') for s in stars.split(',')] * stars_importance)
        soup = ' '.join(filter(None, [soup, stars_soup]))
    
    # Genres.
    genres = x['genre']
    if genres is not np.nan:
        # Almost the same treatment as authors (strip spaces to make matching a bit more likely).
        genre_soup = ' '.join([g.lower().replace(' ', '') for g in genres.split(',')] * genre_importance)
        soup = ' '.join(filter(None, [soup, genre_soup]))
    
    return soup

df['soup'] = df.apply(create_soup, axis=1)
```

```python
df.soup.head()
```

```python
count_vec = CountVectorizer()
count_matrix = count_vec.fit_transform(df['soup'])

cos_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
title_to_index = pd.Series(df.index, index=df['title'])

def get_recommendation(title):
    idx = title_to_index[title]
    scores = pd.Series(cos_sim[idx]).sort_values(ascending=False)
    show_indices = list(scores.iloc[1:11].index)
    return df.iloc[show_indices]

get_recommendation('Sherlock')
```
